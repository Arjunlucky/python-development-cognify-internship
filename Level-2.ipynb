```python
import math
import random
l=int(input('Enter Lower Bound:'))
u=int(input('Enter upper bound:'))
a=random.randint(l,u)

print("\n you've only chance",round(math.log(u - l + 1,2)),"chance to guess the integer:\n")

count=0

while count < math.log(u - l + 1,2):
  count +=1

  guess=int(input(" guess a number:"))
  if a == guess:
        print("Congratulations you did it in ",count, " try")
        break
  elif a > guess:
        print("You guessed too small!")
  elif a < guess:
        print("You Guessed too high!")


if count >= math.log(u - l + 1, 2):
  print("\nThe number is %d" % a)
  print("\tBetter Luck Next time!")
```

```python
import random

n = random.randrange(1,10)
guess = int(input("Enter any number: "))
while n!= guess:
  if guess < n:
    print("Too low")
    guess = int(input("Enter number again: "))
  elif guess > n:
    print("Too high")
    guess = int(input("Enter number again: "))
  else:
    break
print("you guessed it right")
```

```python
import re

def validate_pass(password):
  if len(password) < 8:
        return False
  if not re.search("[a-z]", password):
        return False
  if not re.search("[A-Z]", password):
        return False
  if not re.search("[0-9]", password):
        return False
  return True

password = "nag@123"
if validate_pass(password):
    print("Valid password")
else:
    print("Invalid password")
```
